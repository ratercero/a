```python
# interpolate the flow data
flowdata['flow'] = flowdata['flow'].interpolate(method='linear')
flowdata = flowdata.drop(flowdata.index[0:29]).reset_index(drop=True)
#smoothing with rolling mean
flowdata['smoothed'] = flowdata.iloc[:,1].rolling(window=60).mean()
#fill the nan values at the beginning using fillna
flowdata['smoothed'] = flowdata['smoothed'].fillna(flowdata['flow'])
# interpolate the nh4 data
NH4_data['nh4'] = NH4_data['nh4'].interpolate(method='linear')
NH4_data = NH4_data.drop(flowdata.index[0:29]).reset_index(drop=True)
#smoothing with rolling mean
NH4_data['smoothed'] = NH4_data.iloc[:,1].rolling(window=60).mean()
#fill the nan values at the beginning using fillna
NH4_data['smoothed'] = NH4_data['smoothed'].fillna(NH4_data['nh4'])
# recalculating the flux with smoothed data
NH4_flux_smoothed = flowdata['smoothed']*NH4_data['smoothed'] # g/hr

```

```python
#Plotting the smoothed data
# plotting zoomed in 
#plot together with different plots
# create figure
plt.figure()
# ========= make first graph - plot flow ======================
ax1=plt.subplot(3,1,1)
plt.plot_date(flowdata['time'],flowdata['smoothed'],color='blue',linestyle="-",marker="", 
              label="flow (m3/s)")
# ========= make second graph - plot ammonia concentration ================
ax2=plt.subplot(3,1,2)
plt.plot_date(NH4_data['time'],NH4_data['smoothed'],color='red',linestyle="-"
              ,marker="", label="NH4 (mg/l)")
plt.legend()
#Version 07-11-25 Page 18 of 24
# ========= make third graph - plot ammonia load ================
ax3=plt.subplot(3,1,3)
plt.plot_date(NH4_data['time'],NH4_flux_smoothed*1e-3,
              color='magenta',linestyle="-",marker="",label="flux (kg/hr)") #plot as [kg/hr]
plt.legend()
# ========== make the graphs nicer (all commands refer to the handle ax)
# fix the legend of the x-axis for all the three graphs – let’s say you want a
# value every 6 hours
xt=pd.date_range(start=min(flowdata['time']),end=max(flowdata['time']),freq="8H"
)
ax1.set_xticks(xt)
ax1.set_xticklabels([])
#…and so on for all the other graphs (ax2, ax3, etc…)
ax2.set_xticks(xt)
ax2.set_xticklabels([])

ax3.set_xticks(xt)
ax3.set_xticklabels(xt.strftime("%m-%d-%H"), rotation=90)
# add grid
ax1.grid(True)
ax2.grid(True)
ax3.grid(True)
#…and so on for all the other graphs (ax2, ax3, etc…)
# add label to y axis
ax1.set_ylabel('flow [m3/hr]')
ax2.set_ylabel('NH4 [mg/l]')
ax3.set_ylabel('NH4 load [kg/hr]')

ax1.set_xlim(pd.Timestamp('2018-06-8'), pd.Timestamp('2018-06-15'))
ax2.set_xlim(pd.Timestamp('2018-06-8'), pd.Timestamp('2018-06-15'))
ax3.set_xlim(pd.Timestamp('2018-06-8'), pd.Timestamp('2018-06-15'))
```

```python
# Model 0
data = flowdata.merge(NH4_data[['time', 'nh4', 'smoothed']], on='time')
data['NH4_load'] = NH4_flux_smoothed
# Get parameters and Run model 0 
dry = data[flowdata['smoothed'] < 5000].copy()

time_idx = pd.DatetimeIndex(dry['time'])
t = time_idx.hour/24 + time_idx.minute/1440  # [0,1)

X = np.column_stack([
    np.ones(len(t)),           # a0
    np.sin(2*np.pi*t),         # a1
    np.cos(2*np.pi*t),         # b1
    np.sin(4*np.pi*t),         # a2
    np.cos(4*np.pi*t)          # b2
])
y = dry['NH4_load']
params, residuals, rank, s = np.linalg.lstsq(X, y, rcond=None)

print("Estimated parameters [a0, a1, b1, a2, b2]:")
print(params)
data['flow']=data['smoothed_x']
Model_resulst = NH4inletModel0(params,data)
```

```python
#plot the results
#Plotting the smoothed data
# plotting zoomed in 
#plot together with different plots
# create figure
plt.figure()
# ========= make first graph - plot ammonia concentration ======================
ax1=plt.subplot(3,1,1)
plt.plot_date(NH4_data['time'],NH4_data['smoothed'],color='blue',linestyle="-",marker="", 
              label="Concentration (g/m3)")
plt.plot_date(Model_resulst['time'], Model_resulst['simNH4conc'], color='green', linestyle="-", 
              marker="",label = "simulated concentration")
plt.legend()
# ========= make second graph - plot ammonia load ================
ax2=plt.subplot(3,1,2)
plt.plot_date(NH4_data['time'],NH4_flux_smoothed,color='red',linestyle="-"
              ,marker="", label="NH4 (g/hr)")
plt.plot_date(Model_resulst['time'], Model_resulst['simNH4load'], linestyle ='-',
              marker="", label="simulated NH4 load")
plt.legend()

# ========== make the graphs nicer (all commands refer to the handle ax)
# fix the legend of the x-axis for all the three graphs – let’s say you want a
# value every 6 hours
xt=pd.date_range(start=min(flowdata['time']),end=max(flowdata['time']),freq="24H"
)
ax1.set_xticks(xt)
ax1.set_xticklabels([])
#…and so on for all the other graphs (ax2, ax3, etc…)
ax2.set_xticks(xt)
ax2.set_xticklabels(xt.strftime("%m-%d-%H"), rotation=90)
# add grid
ax1.grid(True)
ax2.grid(True)
#…and so on for all the other graphs (ax2, ax3, etc…)
# add label to y axis
ax1.set_ylabel('NH4 [g/m3]')
ax2.set_ylabel('NH4 load [g/hr]')

ax1.set_ylim([0,100])
#ax1.set_xlim(pd.Timestamp('2018-06-8'), pd.Timestamp('2018-06-15'))
#ax2.set_xlim(pd.Timestamp('2018-06-8'), pd.Timestamp('2018-06-15'))
#ax3.set_xlim(pd.Timestamp('2018-06-8'), pd.Timestamp('2018-06-15'))
```
